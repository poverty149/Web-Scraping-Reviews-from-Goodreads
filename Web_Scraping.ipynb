```python
###########Extracting individual book details and their reviews########################


###Upon receiving an error in this section,measure the length of the list titles using the cell below. Substitute the value in to the range of the for loop until all 45 page_links have been extracted successfully.
for i in range(0,45):
  url=page_links[i]
  x=requests.get(url)
  soup1=BeautifulSoup(x.content,'html.parser')
  title=soup1.find('h1',{'id':'bookTitle'})
  title=title.string.strip()
  num_rating=int(soup1.find('meta',{'itemprop':'ratingCount'}).get('content'))
  num_review=int(soup1.find('meta',{'itemprop':'reviewCount'}).get('content'))
  avg_rating=float(soup1.find('span',{'itemprop':'ratingValue'}).text.strip())
  reviews_text=soup1.findAll('div',{'class':'reviewText stacked'})
  reviews=[]
  for r in reviews_text:
    if(r.findChild('span',{'style':'display:none'})):
      reviews.append(r.findChild('span',{'style':'display:none'}).text)
  titles.append(title)
  num_ratings.append(num_rating)
  num_reviews.append(num_review)
  avg_ratings.append(avg_rating)
  complete_reviews.append(reviews)
```

## Review Collage
 Create a Collage with the most used words in the reviews extracted from the previous section.

```python
import string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk import PorterStemmer
```

```python
STOPWORDS = set(stopwords.words('english'))
```

```python

def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii') # A function to remove emojis from the reviews
```

```python
complete_updated=[]
```

```python
import string   
import re

def clean_text(text):
    ps=PorterStemmer()
    
    text=deEmojify(text) # remove emojis
    text_cleaned="".join([x for x in text if x not in string.punctuation]) # remove punctuation
    
    text_cleaned=re.sub(' +', ' ', text_cleaned) # remove extra white spaces
    text_cleaned=text_cleaned.lower() # converting to lowercase
    tokens=text_cleaned.split(" ")
    tokens=[token for token in tokens if token not in STOPWORDS] # Taking only those words which are not stopwords
    text_cleaned=" ".join([ps.stem(token) for token in tokens])
    
    
    return text_cleaned
```

```python
for revs in complete_reviews:
  updated=[]
  for rev in revs:
    updated.append(clean_text(rev))
  complete_updated.append(updated)
```

```python
s=""
for revs in complete_reviews:
  for rev in revs:
    s+=clean_text(rev)
  
```
